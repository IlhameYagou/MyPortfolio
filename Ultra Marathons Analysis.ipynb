```python
#2.Are runners slower in summer than winter?
#Summer= 6-8
#Winter= 12-2
```

```python
UM2['race_month'] = UM2['race date'].str.split('.').str.get(1).astype(int)
```

```python
UM2.head()
```

```python
UM2['race_season'] = UM2['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter' )
```

```python
UM2.head()
```

```python
UM2.groupby('race_season')['average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
